# Stability-eficncy trade off and VOI

```julia
include("../examples/SurplusProduction.jl")
include("../src/BeliefStateTransitions.jl")
include("../src/MDPSolver.jl")
```

```julia
MSY=10
Fmsy=0.2
SigmaN=0.05
sigma_a=0.5
sigma_p=2.5
price = 1.0
c1=0.0
c2=0.01
c3 = 0.0
low_harvest_penelty = 0.0
BMSY = MSY/Fmsy
b= 1/(BMSY)
maxV = 10.0
discount = 0.05
model_c3_3 = SurplusProduction.init_model3(MSY,Fmsy,SigmaN,sigma_a,sigma_p,c1,
                        c2,c3,low_harvest_penelty, b,maxV,discount;N=70,CVmax=1.0,
                        harvest=collect(0.0:1.0:50),observations = [2])
print(" ")
```

```julia
using Plots
using LaTeXStrings
using Colors
sigma = [log(0.05^2+1), log(0.25^2+1), log(0.5^2+1), log(0.75^2+1)]

meanNodes = model_c3_3.grid.nodesMean

Bt = exp.(meanNodes)
Bt1 = broadcast(x->exp(model_c3_3.mod.T([x],model_c3_3.Policy([x,sigma[1]]),1)[1]),meanNodes)
Plots.plot(Bt ,Bt1, label = "CV = 0.1",c = Colors.RGB(0.3,0.7,0.9), ylab = L"Esc", 
            legend = :topleft,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2)
Bt1 = broadcast(x->exp(model_c3_3.mod.T([x],model_c3_3.Policy([x,sigma[2]]),1)[1]),meanNodes)
Plots.plot!(Bt ,Bt1, label = "CV = 0.25",c = Colors.RGB(0.5,0.7,0.7), ylab = L"Esc", 
            legend = :topleft,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2)
Bt1 = broadcast(x->exp(model_c3_3.mod.T([x],model_c3_3.Policy([x,sigma[3]]),1)[1]),meanNodes)
Plots.plot!(Bt ,Bt1, label = "CV = 0.5",c = Colors.RGB(0.7,0.7,0.5), ylab = L"Esc", 
            legend = :topleft,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2)
Bt1 = broadcast(x->exp(model_c3_3.mod.T([x],model_c3_3.Policy([x,sigma[4]]),1)[1]),meanNodes)
Plots.plot!(Bt ,Bt1, label = "CV = 0.75",c = Colors.RGB(0.9,0.7,0.3), ylab = L"Esc", 
            legend = :topleft,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2)
Plots.plot!([0,100],[0,100],color="black",linestyle=:dash, xlims = [0,100], ylims = [0,100],
ylab = L"B_{t+1}", xlab = L"B_t", title = "Escapement")
```

```julia
Ht1 = broadcast(x->model_c3_3.Policy([x,sigma[1]])[1],meanNodes) 
Plots.plot(Bt,Ht1, label = "CV = 0.1",c = Colors.RGB(0.3,0.7,0.9), ylab = L"H_t", 
            legend = :topleft,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2)
Ht2 = broadcast(x->model_c3_3.Policy([x,sigma[2]])[1],meanNodes)
Plots.plot!(Bt,Ht2, label = "CV = 0.25",c = Colors.RGB(0.5,0.7,0.7), ylab = L"H_t", 
            legend = :topleft,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2)
Ht3 = broadcast(x->model_c3_3.Policy([x,sigma[3]])[1],meanNodes)
Plots.plot!(Bt,Ht3, label = "CV = 0.5",c = Colors.RGB(0.7,0.7,0.5), ylab =  L"H_t", 
            legend = :bottomright,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2, 
            title = "Harvest", xlims = [0,100])
Ht4 = broadcast(x->model_c3_3.Policy([x,sigma[4]])[1],meanNodes)
Plots.plot!(Bt,Ht4, label = "CV = 0.75",c = Colors.RGB(0.9,0.7,0.3), ylab =  L"H_t", 
            legend = :bottomright,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2, 
            title = "Harvest", xlims = [0,100])
```

```julia
Ft1 = Ht1./Bt
Plots.plot(Bt,Ft1, label = "CV = 0.1",c = Colors.RGB(0.3,0.7,0.9), ylab = L"F_t", 
            legend = :topleft,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2)
Ft2 = Ht2./Bt
Plots.plot!(Bt,Ft2, label = "CV = 0.25",c = Colors.RGB(0.5,0.7,0.7), ylab = L"F_t", 
            legend = :topleft,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2)
Ft3 = Ht3./Bt
Plots.plot!(Bt,Ft3, label = "CV = 0.5",c = Colors.RGB(0.7,0.7,0.5), ylab =  L"F_t", 
            legend = :bottomright,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2, 
            title = "effort", xlims = [0,100])
Ft4 = Ht4./Bt
Plots.plot!(Bt,Ft4, label = "CV = 0.75",c = Colors.RGB(0.9,0.7,0.3), ylab =  L"F_t", 
            legend = :bottomright,tickfontsize = 14,guidefontsize=18,legendfontsize=14,linewidth=2, 
            title = "effort", xlims = [0,100])
```

# Load data 

```julia
using JLD2
using DataFrames
file = "../FARM/data/Solution_1_1.csv"
function load_data(file)
    # load file 
    @load  file sol
    
    # add parameters to sims
    dat = sol.sims
    dat.sigma_a = repeat([sol.pars[4]],11)
    dat.C2 = repeat([sol.pars[7]],11)
    dat.C3 = repeat([sol.pars[8]],11)
    dat.ncv = repeat([sol.pars[10]],11)
    
    # compute policy
    CV = collect(0.1:0.1:1.0)
    sigma = log.(CV.^2 .+1)
    meanNodes = log(0.5):0.005:log(110)
    results = DataFrame(CV=[0],mean=[0],sigma_a=[0],c2=[0],c3=[0],ncv=[0],
                        harvest=[0],monitor=[0])
    N = length(meanNodes)
    for i in 1:length(sigma)
        harvest = broadcast(x->sol.P([x,sigma[i]])[1], meanNodes)
        monitor = broadcast(x->sol.P([x,sigma[i]])[2], meanNodes)
        results_i = DataFrame(CV=repeat([CV[i]],N),mean=exp.(meanNodes .+ 0.5*sigma[i]),
                              sigma_a=repeat([sol.pars[4]],N),c2=repeat([sol.pars[7]],N),
                              c3=repeat([sol.pars[8]],N),ncv=repeat([sol.pars[10]],N),
                              harvest = harvest, monitor = monitor)
        results = [results;results_i]
    end
    return results, dat
end
```

```julia
file = "../FARM/data/Solution_1_1.csv"
results, dat = load_data(file)
for i in 1:3
    for j in 1:18
        file = string("../FARM/data/Solution_",i, "_",j,".csv")
        results_ij, dat_ij = load_data(file)
        results = [results;results_ij]
        dat = [dat;dat_ij]
    end
end 
using CSV
CSV.write("../FARM/data/simulations.csv", dat)
CSV.write("../FARM/data/policies.csv", results)
```

```julia
function sim_stats(model)
    N=50
    mu= zeros(N);muB = zeros(N)
    CV= zeros(N);CVB = zeros(N)
    p_t = zeros(N);monitor=zeros(N)
    models = broadcast(i->deepcopy(model),1:Threads.nthreads())
    
    Threads.@threads for i in 1:N
        x0 = [log(MSY/Fmsy)];s0 = ([log(MSY/Fmsy)], [0.3;;]);T = 5000
        quad = BeliefStateTransitions.MvGaussHermite.init_mutable(10,s0[1],s0[2])
        dat=BeliefStateTransitions.simulation_kf(x0,s0,T,filter,models[Threads.threadid()].Policy,
                                                models[Threads.threadid()].mod,models[Threads.threadid()].Returns,quad)
        
        H = broadcast(i -> dat[3][i][1], 1:T)
        B = broadcast(i -> dat[1][i][1], 1:T)

        mu[i] = 1/N*sum(H)/T
        muB[i] = 1/N*sum(exp.(B))/T

        CV[i] =1/N*sqrt(sum(  (H.-N*mu[i]).^2  )/T)/(N*mu[i])
      
        CVB[i]=1/N*sqrt(sum((exp.(B).-N*muB[i]).^2)/T)/(N*muB[i])


        p_t[i]= 1/N*sum(exp.(B) .< 0.75*MSY/Fmsy)/T
        monitor[i]=1/N*sum(broadcast(i -> dat[3][i][2], 1:T).-1)/T
    end 
    
    return [sum(mu),sum(CV),sum(p_t),sum(muB),sum(CVB),sum(monitor)]
end
```

```julia

function mean_var(MSY,Fmsy,SigmaN,sigma_a,sigma_p,p,c3,discount)
    x0 = [log(MSY/Fmsy)]
    s0 = ([log(MSY/Fmsy)], [0.5;;])
    T = 10000

    model = SurplusProduction.init_model2(MSY,Fmsy,SigmaN,sigma_a,sigma_p,p,c3,discount;N=50,CVmax = 0.5)
    quad = BeliefStateTransitions.MvGaussHermite.init_mutable(10,s0[1],s0[2])
    dat=BeliefStateTransitions.simulation_kf(x0,s0,T,filter,model.Policy,model.mod,model.Returns,quad)
    
    H = broadcast(i -> dat[3][i][1], 1:T)
    mu = sum(H)/T
    return mu, sqrt(sum((H.-mu).^2)/T)/mu
end 


function mean_var2(MSY,Fmsy,SigmaN,sigma_a,sigma_p,p,c3,discount)
    x0 = [log(MSY/Fmsy)]
    s0 = ([log(MSY/Fmsy)], [0.5;;])

    model = SurplusProduction.init_model2(MSY,Fmsy,SigmaN,sigma_a,sigma_p,p,c3,discount;N=30,CVmax = 1.0)
    models = broadcast(i->deepcopy(model),1:Threads.nthreads())
    N = 200
    mu= zeros(N);muB = zeros(N)
    CV= zeros(N);CVB = zeros(N)
    p_t = zeros(N)
    Threads.@threads for i in 1:N
        x0 = [log(MSY/Fmsy)];s0 = ([log(MSY/Fmsy)], [0.3;;]);T = 1000
        quad = BeliefStateTransitions.MvGaussHermite.init_mutable(10,s0[1],s0[2])
        dat=BeliefStateTransitions.simulation_kf(x0,s0,T,filter,models[Threads.threadid()].Policy,
                                                models[Threads.threadid()].mod,models[Threads.threadid()].Returns,quad)
        
        H = broadcast(i -> dat[3][i][1], 1:T)
        B = broadcast(i -> dat[1][i][1], 1:T)
        mu[i] = 1/N*sum(H)/T
        muB[i] = 1/N*sum(B)/T
        CV[i] = 1/N*sqrt(sum((H.-N*mu[i]).^2)/T)/(N*mu[i])
        CVB[i] = 1/N*sqrt(sum((B.-N*muB[i]).^2)/T)/(N*mub[i])
        p_t[i]= 1/N*sum(exp.(B) .< 0.75*MSY/Fmsy)/T
    end 
    
    return sum(mu),sum(CV),sum(p_t), sum(muB),sum(CVB)
end

function mean_var3(MSY,Fmsy,SigmaN,sigma_a,sigma_p,c1,c2,c3,b,maxV,discount)
    x0 = [log(MSY/Fmsy)]
    s0 = ([log(MSY/Fmsy)], [0.5;;])
    
    model=SurplusProduction.init_model3(MSY,Fmsy,SigmaN,sigma_a,sigma_p,c1,c2,c3,b,maxV,discount;N=30,CVmax=1.0)

    models = broadcast(i->deepcopy(model),1:Threads.nthreads())
    N = 50
    mu= zeros(N);muB = zeros(N)
    CV= zeros(N);CVB = zeros(N)
    p_t = zeros(N)
    Threads.@threads for i in 1:N
        x0 = [log(MSY/Fmsy)];s0 = ([log(MSY/Fmsy)], [0.3;;]);T = 2000
        quad = BeliefStateTransitions.MvGaussHermite.init_mutable(10,s0[1],s0[2])
        dat=BeliefStateTransitions.simulation_kf(x0,s0,T,filter,models[Threads.threadid()].Policy,
                                                models[Threads.threadid()].mod,models[Threads.threadid()].Returns,quad)
        
        H = broadcast(i -> dat[3][i][1], 1:T)
        B = broadcast(i -> dat[1][i][1], 1:T)
        mu[i] = 1/N*sum(H)/T
        CV[i] = 1/N*sqrt(sum((H.-N*mu[i]).^2)/T)/(N*mu[i])
        muB[i] = 1/N*sum(B)/T
        CVB[i] = 1/N*sqrt(sum((B.-N*muB[i]).^2)/T)/(N*muB[i])
        p_t[i]= 1/N*sum(exp.(B) .< 0.75*MSY/Fmsy)/T
    end 
    
    return sum(mu),sum(CV),sum(p_t), sum(muB),sum(CVB)
end
```

```julia
bhatls = 1:100
sigmals = 0.01:0.0025:0.6
acc1 = zeros(length(bhatls),length(sigmals))
acc2 = zeros(length(bhatls),length(sigmals))
acc3 = zeros(length(bhatls),length(sigmals))
for i in 1:length(bhatls)
    for j in 1:length(sigmals)
        acc1[i,j] = mod3.Policy([log(bhatls[i]),sigmals[j]]) .- mod1.Policy([log(bhatls[i]),sigmals[j]])
        acc2[i,j] = mod3.Policy([log(bhatls[i]),sigmals[j]]) .- mod2.Policy([log(bhatls[i]),sigmals[j]])
        acc3[i,j] = mod2.Policy([log(bhatls[i]),sigmals[j]]) .- mod1.Policy([log(bhatls[i]),sigmals[j]])
    end
end 
p1=Plots.heatmap(sigmals,bhatls,acc1)
p2=Plots.heatmap(sigmals,bhatls,acc2)
p3=Plots.heatmap(sigmals,bhatls,acc3)
plot(p1,p2,p3)
```
