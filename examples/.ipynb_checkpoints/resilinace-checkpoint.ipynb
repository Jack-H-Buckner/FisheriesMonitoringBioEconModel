```julia
include("../src/ValueFunctions.jl")
include("../src/BeliefStateTransitions.jl")
include("SurplusProduction.jl")
include("../src/MDPsolver.jl")
using Plots
using LaTeXStrings
```

```julia
params = SurplusProduction.reparam(10,0.2)
T = (x,actions,aux) -> SurplusProduction.Bt_alt_pstar(x,actions,aux,params)
SigmaN = [0.05;;]

# pstar sigma
Fmin = 0.001
Fmax = 0.2
Bmin = 10.0
Bcrit = 25.0
Hmax = 500
aux = state -> SurplusProduction.pstar_sigma(0.4,state,(Bmin,Bcrit,Fmin,Fmax),Hmax)



# new observation model 
H = [1.0;;]
function SigmaO(actions,aux)
    Sigma_O = zeros(1,1)
    if actions == 1
        Sigma_O[1,1] = 0.1
    else
        Sigma_O[1,1] = 1.5
    end
    return Sigma_O
end 

# define model object to compute grid 

actions = [0,1]
model3=BeliefStateTransitions.init_model(T,aux,H,actions,SigmaO,SigmaN)  


# Define value function 
Bmax = 150.0
Bmin = 1.0
N = 75
CVmax = 0.75
V = ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)


# value function and grid 
grid3=BeliefStateTransitions.init_transitions("quadrature",model3,V;mQuad = 20)

using JLD2
BeliefStateTransitions.computeTransitions!(grid3)
@save "data/grid_3.jld2" grid3
```

```julia
# optimize for differnt monitoring costs 
# c_monitor = 0.5
R1 = (x,actions,aux) -> SurplusProduction.Rpstar(x,actions,aux,10.0,0.0085,0.5) 

rewards1 = BeliefStateTransitions.init_rewards(R1,model3,V)
BeliefStateTransitions.computeRewards!(rewards1)

V1 = MDPsolver.ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)
print(" ")

MDPsolver.solve_parallel(grid3,V1,rewards1,0.9524,verbos = false)

# c_monitor = 1.0
R2 = (x,actions,aux) -> SurplusProduction.Rpstar(x,actions,aux,10.0,0.0085,1.0) 

rewards2 = BeliefStateTransitions.init_rewards(R2,model3,V)
BeliefStateTransitions.computeRewards!(rewards2)

V2 = MDPsolver.ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)
print(" ")

MDPsolver.solve_parallel(grid3,V2,rewards2,0.9524,verbos = false)

# c_monitor = 1.5
R3 = (x,actions,aux) -> SurplusProduction.Rpstar(x,actions,aux,10.0,0.0085,1.5) 

rewards3 = BeliefStateTransitions.init_rewards(R3,model3,V)
BeliefStateTransitions.computeRewards!(rewards3)

V3 = MDPsolver.ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)
print(" ")

MDPsolver.solve_parallel(grid3,V3,rewards3,0.9524,verbos = false)


# c_monitor = 2.0
R4 = (x,actions,aux) -> SurplusProduction.Rpstar(x,actions,aux,10.0,0.0085,2.0) 

rewards4 = BeliefStateTransitions.init_rewards(R4,model3,V)
BeliefStateTransitions.computeRewards!(rewards4)

V4 = MDPsolver.ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)
print(" ")

MDPsolver.solve_parallel(grid3,V4,rewards4,0.9524,verbos = false)
```

```julia

P1=MDPsolver.ValueFunctions.init_Norm2DGrid_obs_policy(log(Bmax),log(Bmin),log(CVmax^2+1),N)
P2=MDPsolver.ValueFunctions.init_Norm2DGrid_obs_policy(log(Bmax),log(Bmin),log(CVmax^2+1),N)
P3=MDPsolver.ValueFunctions.init_Norm2DGrid_obs_policy(log(Bmax),log(Bmin),log(CVmax^2+1),N)
P4=MDPsolver.ValueFunctions.init_Norm2DGrid_obs_policy(log(Bmax),log(Bmin),log(CVmax^2+1),N)
MDPsolver.policy_parallel!(P1,grid3,V1,rewards1,0.9524)
MDPsolver.policy_parallel!(P2,grid3,V2,rewards2,0.9524)
MDPsolver.policy_parallel!(P3,grid3,V3,rewards3,0.9524)
MDPsolver.policy_parallel!(P4,grid3,V4,rewards4,0.9524)
print(" ")
```

```julia
T = 100000
include("../src/MDPsolver.jl")
dat1=MDPsolver.simulation(x0,s0,T,P1,model3,rewards1)
dat2=MDPsolver.simulation(x0,s0,T,P2,model3,rewards2)
dat3=MDPsolver.simulation(x0,s0,T,P3,model3,rewards3)
dat4=MDPsolver.simulation(x0,s0,T,P4,model3,rewards4)
print(" ")
```

```julia
p_Bcrit = [sum(broadcast(i -> exp(dat1[1][i][1]), 1:T) .< Bcrit)/T,
sum(broadcast(i -> exp(dat2[1][i][1]), 1:T) .< Bcrit)/T,
sum(broadcast(i -> exp(dat3[1][i][1]), 1:T) .< Bcrit)/T,
sum(broadcast(i -> exp(dat4[1][i][1]), 1:T) .< Bcrit)/T]

Hcrt = 0.1*25
p_Hcrit = [sum(broadcast(i -> dat1[4][i], 1:T) .< Hcrt)/T,
    sum(broadcast(i -> dat2[4][i], 1:T) .< Hcrt)/T,
    sum(broadcast(i -> dat3[4][i], 1:T) .< Hcrt)/T,
    sum(broadcast(i -> dat4[4][i], 1:T) .< Hcrt)/T
]
c_monitor = [0.5,1.0,1.5,2.0]
Plots.plot(c_monitor,p_Bcrit, label = "")
Plots.scatter!(c_monitor,p_Bcrit, c=1, label = string(L"V =", "biomass"))
Plots.plot!(c_monitor,p_Hcrit,c=2, label = "")
Plots.scatter!(c_monitor,p_Hcrit, c=2, label = string(L"V =", "harvest"), legend = :bottomright,
xlab = string("Cost of monitoring  ", L"(C_{monitor})"), ylab = L"P(V < V_{critical})")
```

```julia
# optimize for differnt monitoring costs 
# c_monitor = 0.5
R1 = (x,actions,aux) -> SurplusProduction.Rpstar_log(x,actions,aux,10.0,0.0085,0.5*log(10)/10) 

rewards1 = BeliefStateTransitions.init_rewards(R1,model3,V)
BeliefStateTransitions.computeRewards!(rewards1)

V1 = MDPsolver.ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)
print(" ")

MDPsolver.solve_parallel(grid3,V1,rewards1,0.9524,verbos = false)

# c_monitor = 1.0
R2 = (x,actions,aux) -> SurplusProduction.Rpstar_log(x,actions,aux,10.0,0.0085,1.0*log(10)/10) 

rewards2 = BeliefStateTransitions.init_rewards(R2,model3,V)
BeliefStateTransitions.computeRewards!(rewards2)

V2 = MDPsolver.ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)
print(" ")

MDPsolver.solve_parallel(grid3,V2,rewards2,0.9524,verbos = false)

# c_monitor = 1.5
R3 = (x,actions,aux) -> SurplusProduction.Rpstar_log(x,actions,aux,10.0,0.0085,1.5*log(10)/10) 

rewards3 = BeliefStateTransitions.init_rewards(R3,model3,V)
BeliefStateTransitions.computeRewards!(rewards3)

V3 = MDPsolver.ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)
print(" ")

MDPsolver.solve_parallel(grid3,V3,rewards3,0.9524,verbos = false)


# c_monitor = 2.0
R4 = (x,actions,aux) -> SurplusProduction.Rpstar_log(x,actions,aux,10.0,0.0085,2.0*log(10)/10) 

rewards4 = BeliefStateTransitions.init_rewards(R4,model3,V)
BeliefStateTransitions.computeRewards!(rewards4)

V4 = MDPsolver.ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)
print(" ")

MDPsolver.solve_parallel(grid3,V4,rewards4,0.9524,verbos = false)
```

```julia

P1=MDPsolver.ValueFunctions.init_Norm2DGrid_obs_policy(log(Bmax),log(Bmin),log(CVmax^2+1),N)
P2=MDPsolver.ValueFunctions.init_Norm2DGrid_obs_policy(log(Bmax),log(Bmin),log(CVmax^2+1),N)
P3=MDPsolver.ValueFunctions.init_Norm2DGrid_obs_policy(log(Bmax),log(Bmin),log(CVmax^2+1),N)
P4=MDPsolver.ValueFunctions.init_Norm2DGrid_obs_policy(log(Bmax),log(Bmin),log(CVmax^2+1),N)
MDPsolver.policy_parallel!(P1,grid3,V1,rewards1,0.9524)
MDPsolver.policy_parallel!(P2,grid3,V2,rewards2,0.9524)
MDPsolver.policy_parallel!(P3,grid3,V3,rewards3,0.9524)
MDPsolver.policy_parallel!(P4,grid3,V4,rewards4,0.9524)
print(" ")
```

```julia
T = 100000
include("../src/MDPsolver.jl")
dat1=MDPsolver.simulation(x0,s0,T,P1,model3,rewards1)
dat2=MDPsolver.simulation(x0,s0,T,P2,model3,rewards2)
dat3=MDPsolver.simulation(x0,s0,T,P3,model3,rewards3)
dat4=MDPsolver.simulation(x0,s0,T,P4,model3,rewards4)
print(" ")
```

```julia
p_Bcrit = [sum(broadcast(i -> exp(dat1[1][i][1]), 1:T) .< Bcrit)/T,
sum(broadcast(i -> exp(dat2[1][i][1]), 1:T) .< Bcrit)/T,
sum(broadcast(i -> exp(dat3[1][i][1]), 1:T) .< Bcrit)/T,
sum(broadcast(i -> exp(dat4[1][i][1]), 1:T) .< Bcrit)/T]

Hcrt = 0.1*25
p_Hcrit = [sum(broadcast(i -> dat1[4][i], 1:T) .< Hcrt)/T,
    sum(broadcast(i -> dat2[4][i], 1:T) .< Hcrt)/T,
    sum(broadcast(i -> dat3[4][i], 1:T) .< Hcrt)/T,
    sum(broadcast(i -> dat4[4][i], 1:T) .< Hcrt)/T
]
c_monitor = [0.1,0.2,0.3,0.4]
Plots.plot(c_monitor,p_Bcrit, label = "")
Plots.scatter!(c_monitor,p_Bcrit, c=1, label = string(L"V =", "biomass"))
Plots.plot!(c_monitor,p_Hcrit,c=2, label = "")
Plots.scatter!(c_monitor,p_Hcrit, c=2, label = string(L"V =", "harvest"), legend = :bottomright,
xlab = string("Cost of monitoring  ", L"(C_{monitor})"), ylab = L"P(V < V_{critical})")
```

# Example

```julia
T = 100
dat=MDPsolver.simulation(x0,s0,T,P1,model3,rewards1)

p1=Plots.plot(broadcast(i -> exp.(dat[2][i][1][1]), 1:T), color = "black", label = "estimate")
Plots.plot!(broadcast(i -> exp.(dat[2][i][1][1] + 2*sqrt(dat[2][i][2][1])), 1:T), 
            fillrange = broadcast(i -> exp.(dat[2][i][1][1] - 2*sqrt(dat[2][i][2][1])), 1:T), 
            alpha =0.2, color = "grey", label = string("uncertainty ", L"(2\sigma)") , ylab = "Biomass")
Plots.scatter!(broadcast(i -> exp(dat[1][i][1]), 1:T), markersize= 2, label = L"B_t")
Plots.plot!(broadcast(i -> exp.(dat[2][i][1][1]- 2*sqrt(dat[2][i][2][1])), 1:T), color = "grey",
            alpha = 0.2, label = "", xlabel = "Time")
p1
```

```julia
biomass = broadcast(i -> exp(dat[1][i][1]), 1:T)
p1=Plots.scatter(biomass, markersize= 3, label = L"B_t", c=3)
Plots.plot!(broadcast(i -> exp(dat[1][i][1]), 1:T),  label = "", color = "black")
Plots.hline!([Bcrit], color = "black", linestyle =:dash, label = "Threshold", ylim = [0,biomass[argmax(biomass)]*1.1])

harvest = broadcast(i -> dat[4][i][1], 1:T)
p2=Plots.scatter(harvest, markersize= 3, label = L"H_t", c=4)
Plots.plot!(harvest, markersize= 2, label = "", color = "black")
Plots.hline!([Hcrt], color = "black", linestyle =:dash, label = "Threshold", ylim = [0,harvest[argmax(harvest)]*1.1])
plot(p1,p2,layout=(2,1))
```

```julia
residual = broadcast(i -> exp.(dat[2][i][1][1]), 1:T).-broadcast(i -> exp(dat[1][i][1]), 1:T)
lag = 0
Plots.plot(residual[1:end-lag], biomass[(1+lag):end])
```

```julia
residual = broadcast(i -> exp.(dat[2][i][1][1]+0.5*dat[2][i][2][1,1]), 1:T).-broadcast(i -> exp(dat[1][i][1]), 1:T)
lag = 0
Plots.plot(residual[1:end-lag], biomass[(1+lag):end])
```

```julia
Plots.scatter(biomass,harvest)
```
