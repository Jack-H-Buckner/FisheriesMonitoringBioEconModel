# Deterministic solutions

```julia
include("../FARM/harvest_policies.jl")
```

```julia
# set parameters
MSY = 10
Fmsy = 0.2
SigmaN = [0.000005;;]
sigma_a = [0.1,0.2,0.4]
sigma_p = 2.0
c1 = 0.0
C2 = [0.0,0.02,0.04]
C3 = [0.0, 3.0, 9.0]
b = Fmsy
maxV = [0.0,MSY]
discount = 0.05
beta = 1/(1+discount)
# set value and policy functions 
pars = (MSY,Fmsy,SigmaN,sigma_a[1],sigma_p,c1,0.01,C3[1],b,maxV[1],discount)
R1, V, P, mod, delta=harvest_policies.set_up_model(pars;N=20)

using ForwardDiff
R = (x,h) -> R1([log(x)],(h,1),1)
T = (x,h) -> exp(mod.T([log(x)],(h,1),1)[1])
dRdh(x,h) = ForwardDiff.derivative(h -> R(x,h),h)
dRdx(x,h) = 1 #ForwardDiff.derivative(x -> R(x,h),x)
dTdh(x,h) = ForwardDiff.derivative(h -> T(x,h),h)
dTdx(x,h) = ForwardDiff.derivative(x -> T(x,h),x)
```

```julia
function target(x;discount=0.05) 
    B=x[1];h=x[2];lambda=x[3];beta = 1/(1+discount)
    FOC = dRdh(B,h) + beta * dTdh(B,h) * lambda
    eq = B - T(B,h) 
    ENV = lambda - beta*lambda - dRdx(B,h)
    return [FOC,eq,ENV]
end 
```

```julia
using NLsolve
nlsolve(x->target(x,discount=0.05),[60,10,0.5],ftol=10^-15)
```
