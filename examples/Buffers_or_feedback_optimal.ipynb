# Buffers or feedback optimal

```julia
include("../examples/SurplusProduction.jl")
include("../src/BeliefStateTransitions.jl")
include("../src/MDPSolver.jl")
```

```julia
MSY=10; Fmsy=0.3
SigmaN=0.05; sigma_a=0.2; sigma_p=2.5
price = 1.0; c1=0.0; c2=0.00; c3 = 0.03
low_harvest_penelty = 0.0
BMSY = MSY/Fmsy
b= 1/(BMSY)
maxV = 0.0
discount = 0.05
model_1= SurplusProduction.init_model3(MSY,Fmsy,SigmaN,sigma_a,sigma_p,c1,
                        c2,c3,low_harvest_penelty, b,maxV,discount;N=100,CVmax=1.0,
                        harvest=collect(0.0:1.0:20),observations = [2])
nothing
```

```julia
Bval = 1:0.01:2*BMSY
using Plots
Plots.plot(Bval,broadcast(B -> model_c3_3.Policy([log(B),0.1])[1],Bval))
Plots.plot!(Bval,broadcast(B -> model_c3_3.Policy([log(B),0.2])[1],Bval))
Plots.plot!(Bval,broadcast(B -> model_c3_3.Policy([log(B),0.3])[1],Bval))
```
