```julia
include("../src/MvGaussHermite.jl")
include("../src/BeliefStateTransitions.jl")
include("../examples/SurplusProduction.jl")
params = [10.,100.]
T = (x,actions,aux) -> SurplusProduction.Bt_alt(x,actions,aux,params)
SigmaN = [0.075;;]


# set observations model
H = [1.0;;]
function SigmaO(actions,aux)
    H,obs=actions
    Sigma_O = zeros(1,1)
    if obs == 1
        Sigma_O[1,1] = 0.1
    else
        Sigma_O[1,1] = 10.0
    end
    return Sigma_O
end 

# pstar sigma
Fmin = 0.02
Fmax = 0.2
Bmin = 0.0
Bcrit = 0.0
Hmax = 200
aux = state -> SurplusProduction.pstar_sigma(0.4,state,(Bmin,Bcrit,Fmin,Fmax),Hmax)


# Set grid for harvest rates and monitoring levels
harvest = collect(0.0:2.5:50)
observations = [0,1]
actions = reshape(collect(Iterators.product(harvest, observations)), length(harvest)*length(observations))
        
# define model object to compute grid 
model=BeliefStateTransitions.init_model(T,aux,H,actions,SigmaO,SigmaN)  


m = 15
quad = MvGaussHermite.init_mutable(m,zeros(size(model.SigmaN)[1]),model.SigmaN)
print(" ")
```

```julia
## simulate time series 
s = ([log(100)],[0.1;;])
x = [log(100)]
H = aux(s)
a = (H,1)
x = T(x,a,1.0)
x = x .+ rand(Distributions.MvNormal(zeros(1),SigmaN))
y = x 
s_prime = BeliefStateTransitions.filter!(y,s,a,model,quad)
```

```julia
N = 5000
samples = reshape(rand(Distributions.MvNormal(s[1],s[2]),N),N)

updated_samples = broadcast(i -> T([samples[i]],a,1)[1], 1:N).+ reshape(rand(Distributions.MvNormal([0.0],SigmaN),N),N)

weights = broadcast(i->pdf(Distributions.MvNormal([0],SigmaO((H,1),H)),y .- updated_samples[i]), 1:N)
weights = weights./sum(weights)

E = sum(updated_samples .*weights)
println("Mean = ", E)
var = sum((updated_samples .-E).^2 .*weights)
println("var = ", var)
```

```julia
function particle_filter(y,s,a,model;N=50000)
    
    samples = reshape(rand(Distributions.MvNormal(s[1],s[2]),N),N)

    updated_samples = broadcast(i -> model.T([samples[i]],a,1)[1], 1:N).+ reshape(rand(Distributions.MvNormal([0.0],model.SigmaN),N),N)

    weights = broadcast(i->pdf(Distributions.MvNormal([0], model.SigmaO(a,H)),y .- updated_samples[i]), 1:N)
    weights = weights./sum(weights)

    E = sum(updated_samples .*weights)
    
    var = sum((updated_samples .-E).^2 .*weights)
    return [E], [var;;]
    
end 

using StatsBase


function particle_filter(s,a,model;N=50000)
    
    samples = reshape(rand(Distributions.MvNormal(s[1],s[2]),N),N)

    updated_samples = broadcast(i -> model.T([samples[i]],a,1)[1], 1:N).+ reshape(rand(Distributions.MvNormal([0.0],model.SigmaN),N),N)

    return updated_samples
    
end 
```

```julia
s = ([log(100)],[0.1;;])
H = aux(s)
a = (H,0)
xprime = model.T([log(150)], a, H)[1]
y = [xprime]

println(particle_filter(y,s,a,model;N=5000))
println(BeliefStateTransitions.filter!(y,s,a,model,quad))
```

# Compare $T'(s,a)$ histograms 

```julia


Bhat = 20:20:100
sig = 0.05:0.1:0.35
N_ = 10000

dat = zeros(length(Bhat)*length(sig)*N_,4)
obs = 0
k=0

for b in Bhat
    print(b, " ")
    for v in sig
        k+=1
        
        s = ([log(b)],[v;;])
        H = aux(s)
        a = (H,obs)

        s_prime = BeliefStateTransitions.filter!(s,a,model,quad)
        samples_kf = reshape(rand(Distributions.MvNormal(s_prime[1],s_prime[2]),N_),N_)
        samples_pf = particle_filter(s,a,model;N=N_)
    
        dat[((k-1)*N_ + 1):(k*N_),1] .= b
        dat[((k-1)*N_ + 1):(k*N_),2] .= v
        dat[((k-1)*N_ + 1):(k*N_),3] = samples_kf
        dat[((k-1)*N_ + 1):(k*N_),4] = samples_pf

    end 
end 

using DelimitedFiles

writedlm( "data/surplus_production_tests_histograms.csv",  dat, ',')
```

```julia
function calcualte_int_pf!(pf,s,a,model)
    BeliefStateTransitions.integrate!(pf,s,a,model)
    return sum(broadcast(i -> V1([pf.intermidiate[i][1][1],pf.intermidiate[i][2][1,1]]), 1:length(pf.weights)))/length(pf.weights)
end 

function calcualte_int_kf!(kf,s,a,model)
    BeliefStateTransitions.integrate!(kf,s,a,model)
    return sum(broadcast(i -> V1([kf.intermidiate[i][1][1],kf.intermidiate[i][2][1,1]]), 1:length(kf.weights_y)).*kf.weights_y)
end 
```

```julia
Bhat = 20:20:100
sig = 0.05:0.1:0.35

acc_pf = zeros(length(Bhat), length(sig))
acc_kf = zeros(length(Bhat), length(sig))
obs=0       
i=0
for b in Bhat
    i+=1;j=0
    print(b, " ")
    for v in sig
        j+=1
        
        s = ([log(b)],[v;;])
        H = aux(s)
        a = (H,obs)

        acc_pf[i,j] = calcualte_int_pf!(pf,s,a,model)
        acc_kf[i,j] = calcualte_int_kf!(kf,s,a,model)

    end 
end 

using DelimitedFiles

writedlm( "data/value_function_int_pf.csv",  acc_pf, ',')
writedlm( "data/value_function_int_kf.csv",  acc_kf, ',')
```

# Compare solutions with particle filter and quadrature 

```julia
include("../src/MvGaussHermite.jl")
include("../src/BeliefStateTransitions.jl")
include("../examples/SurplusProduction.jl")
include("../src/ValueFunctions.jl")
include("../src/MDPSolver.jl")

params = SurplusProduction.reparam(10,0.2)
T = (x,actions,aux) -> SurplusProduction.Bt_alt_pstar(x,actions,aux,params)
SigmaN = [0.05;;]

# pstar sigma
Fmin = 0.02
Fmax = 0.2
Bmin = 0.0
Bcrit = 0.0
Hmax = 200
aux = state -> SurplusProduction.pstar_sigma(0.4,state,(Bmin,Bcrit,Fmin,Fmax),Hmax)

# new observation model
H = [1.0;;]
function SigmaO(actions,aux)
    Sigma_O = zeros(1,1)
    if actions == 1
        Sigma_O[1,1] = 0.1
    else
        Sigma_O[1,1] = 1.5
    end
    return Sigma_O
end 

# define model object to compute grid 
actions = [0,1]
model3=BeliefStateTransitions.init_model(T,aux,H,actions,SigmaO,SigmaN)  


# Define value function 
Bmax = 150.0
Bmin = 1.0
N = 75
CVmax = 0.75
V = ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)



grid3=BeliefStateTransitions.init_transitions("particle filter",model3,V;Nfilter=5000,NMC = 100)

print(" ")
```

```julia
BeliefStateTransitions.computeTransitions!(grid3)
```

```julia
R5 = (x,actions,aux) -> SurplusProduction.Rpstar(x,actions,aux,10.0,0.0085,0.1) 

rewards5 = BeliefStateTransitions.init_rewards(R5,model3,V)
BeliefStateTransitions.computeRewards!(rewards5)

V5 = MDPsolver.ValueFunctions.init_Norm2DGrid(log(Bmax),log(Bmin),log(CVmax^2+1),N)
print(" ")

MDPsolver.solve_parallel(grid3,V5,rewards5,0.9524)
```

```julia

N = 75
values_1 = zeros(N,N)
values_0 = zeros(N,N)
for i in 1:N
    for j in 1:N
        values_1[i,j]=MDPsolver.bellman!(zeros(2),2,i,j,grid3,V5,rewards5,0.9524)
        values_0[i,j]=MDPsolver.bellman!(zeros(2),1,i,j,grid3,V5,rewards5,0.9524)
    end
end 

Plots.heatmap(grid3.nodesVar,exp.(grid3.nodesMean),values_1.-values_0)
Plots.contour!(grid3.nodesVar,exp.(grid3.nodesMean),values_1.-values_0, color = "white", levels = [0.0])
```
